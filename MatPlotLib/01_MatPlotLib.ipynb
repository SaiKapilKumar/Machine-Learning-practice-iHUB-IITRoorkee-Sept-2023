# Matplotlib

Matplotlib is one of the most popular libraries for creating data visualizations. It is often compared to MATLAB.

**Pros:**
1. Easy to get started.
2. Supports custom labels and texts.
3. Provides control over every element in a figure.
4. Highly customizable.

**Cons:**
1. Steeper learning curve for complex visualizations.
2. Default styles may not always produce aesthetically pleasing plots.
3. Limited support for interactive plots (although libraries like Plotly can supplement this).


```python
import matplotlib.pyplot as plt

#%matplotlib inline
#plt.show() : Similar to Print Statement
```

```python
import numpy as np
import pandas as pd
import math
```

```python
x = np.linspace(0,10,100)
y = x**2
z = np.sqrt(x)
```

# Basic MatPlotLib Commands

```python
dir(plt)
```

```python
plt.plot(x,y,'red')
plt.xlabel("Put X-axis title here")
plt.ylabel("Put Y-axis title here")
plt.title('Put Title for this figure')
plt.show()
```

# Object Oriented Method

```python
fig = plt.figure() #empty canvas

axes = fig.add_axes([0.11,0.1,0.8,0.8])
axes2 = fig.add_axes([1.1,.1,0.8,0.8])

axes.plot(x,y,'r',label ='x**2')
axes2.plot(x,z,'b',label ='sqrt(x)')

axes.set_xlabel('Set X label here')
axes.set_ylabel('Set Y label here')
axes.set_title('Set Title')

axes2.set_xlabel('Set X label here')
axes2.set_ylabel('Set Y label here')
axes2.set_title('Set Title')
plt.show()
#left, bottom, width, height
```

```python
j = x**3
```

```python
fig = plt.figure()

axes3 = fig.add_axes([0,0,1,1])
axes3.plot(x,y, label= 'x**2')
axes3.plot(x,j, label= 'x**3')
axes3.legend()
plt.show()
```
