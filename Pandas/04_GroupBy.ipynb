# GroupBy

```python
import numpy as np
import pandas as  pd

```

```python
df = pd.DataFrame({
    'Company':['Apple','Apple','Google','Google','Meta','Meta'],
    'Emp':['Mohit','Ganeshan','Vanessa','Sara','Vikram','Pragyan'],
    'Sales':[100,120,210,187,342,290]
})

df
```

```python
df.groupby('Company')
```

```python
df.groupby('Company').mean(numeric_only = True)
```

```python
df.groupby('Company').max()
```

```python
df.groupby('Company').sum(numeric_only = True)
```

```python
by_company = df.groupby('Company')
```

```python
by_company.std(numeric_only = True)
```

```python
by_company.describe()
```

```python
by_company.describe().transpose()
```
