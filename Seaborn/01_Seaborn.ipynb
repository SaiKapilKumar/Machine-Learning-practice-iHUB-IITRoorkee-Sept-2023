# Seaborn

statistical plotting library built on top of matplotlib

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = pd.read_csv('tips.csv')
```

```python
df
```

# Distribution plot

shows distribution of a univariate(one variable) set of observations.

```python
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
sns.distplot(df['total_bill'],kde = True)
```

kde: kernel density estimation calculates the density of features in the neighbourhood around those features.

```python
sns.distplot(df['total_bill'],kde = True, bins = 30)



#bins defines smoothness or sharpness in your plot
```

# jointplot

match up two distplots for bivariate data. kind parameter: 'scatter', 'reg', 'kde', 'hex', 'resid'

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind= 'reg')
plt.show()
```

```python
df[df['total_bill']>46]
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind= 'kde')
plt.show()
```

# Categorical Data Plots

### Bar plot and Count Plot

```python
sns.barplot(x='day',y='tip',data=df, hue = 'sex', estimator=np.mean)
```

```python
sns.barplot(x='day',y='tip',data=df, hue = 'sex')
```

```python
sns.countplot(data=df, x='smoker')
```

# Boxplot and Violinplot

```python
sns.violinplot(y='smoker',x='total_bill',data = df, orient = 'h')
```
